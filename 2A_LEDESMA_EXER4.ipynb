1. **Data Explorations**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
print(data.head)
```

Each feature/data will have an effect or impact on the final grades. The grades will be high if the datas each section are high and vice versa. Another is if there is a missing data on some section, it will show in the grade that you missed something

2. **Data Preprocessing**

```python
print(data.isnull().sum())
```

Standardazing and normalizing numerical features is important because it helps you realize if there are some data that are missing. You will be able to identify them and fill the missing data if needed

3. **Train-Test Split**

```python
x = data.drop('Grades', axis=1)
y = data['Grades']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

4. **Linear Regression Model**

```python
plt.scatter(data['Hours_of_Study'], data['Grades'])
plt.xlabel('Hours of Study')
plt.ylabel('Grades')
plt.title('Scatter Plot of Data #1')
plt.show()
```

```python
plt.scatter(data['Attendance_Percentage'], data['Grades'])
plt.xlabel('Attendance Percentage')
plt.ylabel('Grades')
plt.title('Scatter Plot of Data #2')
plt.show()
```

```python
plt.scatter(data['Previous_Exam_Scores'], data['Grades'])
plt.xlabel('Previous Exam Scores')
plt.ylabel('Grades')
plt.title('Scatter Plot of Data #3')
plt.show()
```

```python
plt.scatter(data['Extracurricular_Activities'], data['Grades'])
plt.xlabel('Extracurricular Activities')
plt.ylabel('Grades')
plt.title('Scatter Plot of Data #4')
plt.show()
```

```python
plt.scatter(data['Travel_Time_to_School'], data['Grades'])
plt.xlabel('Travel Time to School')
plt.ylabel('Grades')
plt.title('Scatter Plot of Data #5')
plt.show()
```

```python
x = data[['Previous_Exam_Scores']]
y = data['Grades']

model = LinearRegression()
model.fit(x,y)

plt.scatter(x,y)
plt.plot(x,model.predict(x),color='Black')
plt.xlabel('Previous_Exam_Scores')
plt.ylabel('Grades')
plt.title('Linear Regression #1')
plt.show()
```

```python
j = data[['Hours_of_Study']]
c = data['Grades']

model = LinearRegression()
model.fit(j,c)

plt.scatter(j,c)
plt.plot(j,model.predict(j),color='pink')
plt.xlabel('Hours of Study')
plt.ylabel('Grades')
plt.title('Linear Regression #2')
plt.show()
```

```python
a = data[['Attendance_Percentage']]
m = data['Grades']

model = LinearRegression()
model.fit(a,m)

plt.scatter(a,m)
plt.plot(a,model.predict(a),color='red')
plt.xlabel('Attendance Percentage')
plt.ylabel('Grades')
plt.title('Linear Regression #3')
plt.show()
```

```python
n = data[['Travel_Time_to_School']]
c = data['Grades']

model = LinearRegression()
model.fit(n,c)

plt.scatter(n,c)
plt.plot(n,model.predict(n),color='violet')
plt.xlabel('Attendance Percentage')
plt.ylabel('Grades')
plt.title('Linear Regression #4')
plt.show()

```
